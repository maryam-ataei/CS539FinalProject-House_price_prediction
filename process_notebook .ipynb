#Overview

This project utilizes Melbourne Airbnb open data to predict the rental price of a short-stay house using machine learning techniques such as data preprocessing, feature selection, model training, and evaluation. Several regression models are applied, including linear regression, Support Vector Machines, decision tree, random forest and Multi Layer Perceptron and these models are evaluated using metrics such as Mean Squared Error and Cross Validation Mean Squared Error.

Overall, the project provides insights into the factors that influence the rental price of a short-stay house and demonstrates the effectiveness of machine learning techniques for predicting rental prices.

#Methods

The projet began with data analysis.
For the data analysis, the first step was to conduct exploratory data analysis (EDA) to gain insights into the dataset's features and identify any outliers or missing values. We used Python's Pandas library and visualization tools such as Matplotlib and Seaborn for EDA. After EDA, we performed data preprocessing, which involved handling missing data, removing outliers, scaling and normalizing the numerical data, and encoding categorical data.
In the data cleaning stage,  we first identified any missing values within the dataset and used the mean values to replace them. In cases where there were only a few missing values for more effective features, we removed them. We also checked for duplicates and removed them if any were found. We then checked for outliers using visualizations and statistical methods such as the Z-score. We also used histogram and also find the proportion of values that fell outside of a certain range. Based on this analysis, we decided to remove any data points that were deemed to be outliers. We also used label encoding and one hot encoding for encoding categorical data and find that label encoding have a better result for us. During data preprocessing, we processed the amenities column which contained text data in the form of sentences. We first converted the text to lowercase and removed special characters such as braces, quotes, parentheses, and colons. Then, using regular expressions, we split the amenities column into individual values and created a list of values for each row. Finally, we created separate columns for each amenity, with a Boolean value indicating whether that amenity was present in that row's list of amenities. Then, we checked for inconsistent or invalid data such as incorrect or inconsistent data types and values, which were also cleaned and standardized. The cleaned dataset was then used for feature engineering and model building. Finally, our cleaned data has 22799 samples for 212 features.


Next, we conducted feature selection to identify the most important features that have the greatest impact on the rental price. We used correlation heatmap and permutation methods to select the most relevant features.

For model training, we utilized five different regression models, including Linear Regression, Support Vector Regression (SVM), Decision Tree Regression, Random Forest Regression and Multi Layer Perceptron (MLP) . 
* Linear Regression: A simple and popular method that models the relationship between the input variables and the output variable as a linear function.

* Decision Trees: A method that builds a tree-like model of decisions and their possible consequences, and predicts the output variable based on the input variables.

* Random Forest: A method that builds an ensemble of decision trees, where each tree is trained on a subset of the input data and a random subset of the input variables.

* Support Vector Machines (SVMs): A method that finds the hyperplane that best separates the input data into two classes (in the case of regression, the classes are above and below the hyperplane).

* Multi Layer Perceptron (MLP): MLP is a type of neural network with multiple layers of neurons that are used in supervised learning tasks such as classification and regression. They are trained using backpropagation and have been applied to various real-world applications.

We trained each model on the preprocessed dataset with the selected features, and evaluated their performance using mean squared error (MSE) and cross-validation mean squared error (CV). 
* Mean Squared Error (MSE) is a common metric used to evaluate the performance of regression models in machine learning. It measures the average squared difference between the predicted and actual values of a dependent variable. The formula for calculating MSE involves taking the sum of the squared differences between each predicted and actual value, and dividing by the total number of observations. A lower MSE indicates better performance, as it means that the model's predictions are closer to the actual values. MSE is widely used in fields such as economics, engineering, and finance to evaluate the accuracy of statistical models.
*Cross-validation (CV) is a technique used in machine learning to evaluate the performance of a model on a dataset. It involves partitioning the dataset into multiple subsets, or "folds", training the model on some of the folds and testing it on the remaining folds. The main purpose of cross-validation is to estimate the performance of a model on new, unseen data and to avoid overfitting. Cross-validation is an important tool for model selection and hyperparameter tuning, as it provides a more accurate estimate of a model's performance than simply using a single training/testing split.

Finally, we selected the random forest regression model as the final model based on its superior performance in terms of accuracy and robustness to outliers. We trained the model on 80% of the dataset and tested it on the remaining 20%, achieving an MSE value of 11296.5279 and CV value of 14551.281 on the test set.

In summary, the methods used for data analysis included EDA, data preprocessing, feature selection, and model training and evaluation. These methods were performed using various Python libraries and techniques to achieve the best results.
